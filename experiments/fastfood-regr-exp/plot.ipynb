```python
import pandas as pd
import numpy as np
import glob
import json
import matplotlib.pyplot as plt
import matplotlib
import matplotlib2tikz 
matplotlib.rc_file('/home/srossi/.config/matplotlib/matplotlibrc')
import sys
import os
sys.path.append("../../")
import vardl
```

```python
json_files = glob.glob('**/results.json', recursive=True)

len(json_files)
```

```python
dfs = []
for file in json_files:
    dfs.append(pd.DataFrame(pd.read_json(file, typ='series')).T)
```

```python
df = pd.concat(dfs).reset_index()
del df['index']
df['spb'] = 1 / (df['total_iters'] / 24 / 3600) # seconds per batch 
df
```

```python
import seaborn as sns
g = sns.catplot(data=df[df['nfeatures'] == 256][df['nlayers'] == 10], x='dataset', y='test_error', kind="bar", hue='model', legend_out=False)
g.set_xticklabels(rotation=30)
plt.legend()
plt.show()
```

```python
g = sns.catplot(data=df, row='nlayers', col='nfeatures', x='dataset', y='test_error', kind="bar", hue='model', height=4, legend_out=False, margin_titles=True, sharey='row')
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(40)
#g.savefig('./figures/test_error.pdf')
#vardl.utils.ExperimentPlotter.savefig('./figures/test_error', 'tex')
plt.show()
```

```python
g = sns.catplot(data=df, row='nlayers', col='nfeatures', x='dataset', y='test_mnll', kind="bar", hue='model', height=4, legend_out=False, margin_titles=True, sharey='row')
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(40)
g.savefig('./figures/test_mnll.pdf')
#vardl.utils.ExperimentPlotter.savefig('./figures/test_mnll', 'tex')
plt.show()
```

```python
g = sns.catplot(data=df, 
                col='dataset',
                hue='model',
                y='test_mnll',
                x='trainable_parameters',
                kind="strip",
                height=4,
                col_wrap=4,
                legend_out=None,
                margin_titles=True,
                sharey=None,)
                
g.set_xticklabels(rotation=90)
for ax in g.axes:
    ax.set_xscale('log')
    #    ax.set_yscale('log')
    
g.savefig('./figures/params_vs_mnll.pdf')
#vardl.utils.ExperimentPlotter.savefig('./figures/params_vs_mnll', 'tex')
```
